```python
import pandas as pd
```

```python
ans=pd.read_csv("loan.csv")
ans
```

```python
ans.isnull().sum()
```

```python
ans['Gender'].fillna(ans['Gender'].mode()[0],inplace=True)
ans['Married'].fillna(ans['Married'].mode()[0],inplace=True)
```

```python
en_data=ans[["Gender","Married"]]
#en_data
```

```python
pd.get_dummies(en_data)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
ar=ohe.fit_transform(en_data).toarray()
```

```python
pd.DataFrame(ar,columns=["Gender_Female","Gender_Male","Married_No","Married_Yes"])
```

```python
ohe=OneHotEncoder(drop="first")
ar1=ohe.fit_transform(en_data).toarray()
ar1
```

```python
pd.DataFrame(ar1,columns=["Gender_Male","Married_Yes"])
```

```python
Ordinal Encoding
```

```python
import pandas as pd
```

```python
aa=pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]})
aa
```

```python
ord_data=[["s","m","l","xl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=ord_data)
oe.fit(aa[["Size"]])
```

```python
aa["Size_en"]=oe.transform(aa[["Size"]])
```

```python
aa
```

```python
ord_data1 = {"s": 5, "m": 6, "l": 7, "xl": 8}
```

```python
aa["size_en_map"]=aa["Size"].map(ord_data1)
```

```python
aa
```
