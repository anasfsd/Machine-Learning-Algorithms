```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset=pd.read_csv("loan.csv")
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
q1=dataset["CoapplicantIncome"].quantile(0.25)
q3=dataset["CoapplicantIncome"].quantile(0.75)
iqr=q3-q1
```

```python
min_r = q1-(1.5*iqr)
max_r = q1+(1.5*iqr)

min_r,max_r
```

```python
sns.distplot(dataset["CoapplicantIncome"])
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=np.log1p)
```

```python
ft.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_a']=ft.transform(dataset[['CoapplicantIncome']])
```

```python
plt.figure(figsize=(15,4))

plt.subplot(1,2,1)
sns.distplot(dataset[['CoapplicantIncome']])
plt.title("Before")

plt.subplot(1,2,2)
sns.distplot(dataset[['CoapplicantIncome_a']])
plt.title("After")
plt.show()
```
