```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("iris.csv")
dataset.drop(columns=["variety"],inplace=True)
```

```python
dataset
```

```python
sns.pairplot(data=dataset)#they work with distance
```

```python
import scipy.cluster.hierarchy as sc
```

```python
plt.figure(figsize=(20,20))
#sc.dendrogram(sc.linkage(dataset,method='complete', metric='euclidean'))
sc.dendrogram(sc.linkage(dataset,method='complete', metric='euclidean'))

plt.savefig("demo.jpg")
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
ac=AgglomerativeClustering(n_clusters=2,linkage='single')
dataset["Predict"]=ac.fit_predict(dataset)
```

```python
dataset.head(3)
```

```python
sns.pairplot(data=dataset,hue="Predict")
plt.show()#they work with distance
```
